```python
def generate_jsons(json_path, key, images, file_name):
    with open(json_path) as f:
        data = json.load(f)
    dic = {}
    dic['info'] = data['info']
    for ch in data[key]:
        if ch['image_id'] in images:
            dic.update(ch)
    with open(file_name + '.json', 'w') as json_file:
        json.dump(dic, json_file)
```

## Plot for similarity

```python
def plot_select(df, groupbycol = 'similarity_rate', similarity_rate = None,
                epoch = None, filters = None, subplots = True):
    
    tdf = df[df['decoder_filter_reverse'] == False]
    if epoch:
        tdf = tdf[tdf['epoch'] == epoch]
    if filters:
        tdf = tdf[tdf['filters'] == filters]
    if similarity_rate:
        tdf = tdf[tdf['similarity_rate'] == similarity_rate]
    
    tdf = tdf[['dataset','epoch','batch','similarity_rate', 'filters','recon','mse','mae','common_nn','rand_index']]
    
    similarity_df = tdf.groupby(groupbycol).mean()
    similarity_df = similarity_df[['recon','mae','common_nn']]
    similarity_df.plot(subplots = subplots)
    print(similarity_df.head())
```
